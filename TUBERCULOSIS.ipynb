```python
plt.figure(figsize=(8,5))

plt.hist(data["Estimated incidence of all forms of tuberculosis"], bins=20)

plt.title("Distribusi Data Kasus Tuberkulosis")
plt.xlabel("Jumlah Kasus")
plt.ylabel("Frekuensi")

plt.show()
```

```python
plt.figure(figsize=(8,5))

sns.boxplot(data=data, y="Estimated incidence of all forms of tuberculosis")

plt.title("Boxplot Data Kasus Tuberkulosis")
plt.ylabel("Jumlah Kasus")

plt.show()
```

```python
plt.figure(figsize=(10,5))

data["Entity"].value_counts().head(10).plot(kind="bar")

plt.title("10 Kategori dengan Jumlah Kasus Tertinggi")
plt.xlabel("Kategori")
plt.ylabel("Jumlah")

plt.xticks(rotation=45)

plt.show()
```

```python
plt.figure(figsize=(10,5))

# Ensure the incidence column is numeric for calculations
data['Estimated incidence of all forms of tuberculosis'] = data['Estimated incidence of all forms of tuberculosis'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
data['Estimated incidence of all forms of tuberculosis'] = pd.to_numeric(data['Estimated incidence of all forms of tuberculosis'], errors='coerce')

data.groupby("Year")["Estimated incidence of all forms of tuberculosis"].sum().plot(kind="line", marker="o")

plt.title("Tren Kasus Tuberkulosis Berdasarkan Tahun")
plt.xlabel("Tahun")
plt.ylabel("Jumlah Kasus")

plt.grid()

plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.heatmap(
    data.corr(numeric_only=True),
    annot=True
)

plt.title("Korelasi Antar Variabel")

plt.show()
```

```python
import matplotlib.pyplot as plt

# Convert the 'Estimated incidence of all forms of tuberculosis' column to numeric
df['Estimated incidence of all forms of tuberculosis'] = df['Estimated incidence of all forms of tuberculosis'].astype(str).str.replace('.', '', regex=False).str.replace(',', '.', regex=False)
df['Estimated incidence of all forms of tuberculosis'] = pd.to_numeric(df['Estimated incidence of all forms of tuberculosis'], errors='coerce')

data_tahun = df.groupby('Year')['Estimated incidence of all forms of tuberculosis'].mean()

plt.figure(figsize=(10,5))
plt.plot(data_tahun.index, data_tahun.values)

plt.xlabel('Tahun')
plt.ylabel('Estimasi Kasus TBC')
plt.title('Tren Estimasi Kejadian Tuberkulosis Tahun ke Tahun')

plt.show()
```
